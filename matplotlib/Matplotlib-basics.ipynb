```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Basic Graph
# for plotting simple graph
x = [5,10,15,20,25]
y = [30,35,40,45,50]

plt.plot(x,y)
plt.show()
```

```python
#color
plt.plot(x,y,color='Yellow')
```

```python
#for marking in the graph
x = [5,10,15,20,25]
y = [30,35,40,45,50]
plt.plot(x,y,color='yellow',marker='*')
plt.show()
```

```python
# sixe and marker of color
plt.plot(x,y,color='yellow',marker='*', markeredgecolor='blue',markersize=10)
```

```python
plt.plot(x,y,color='yellow',marker='*',linestyle='-.', markeredgecolor='blue',markersize=10)
```

```python
# shortform(color,marker,linestyle)
plt.plot(x,y,'r*--')
```

```python
# for heading of the graph
plt.plot(x,y,'r*--')
plt.title('Heading1',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
```

```python
# For labeling the axis
plt.plot(x,y,'r*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('First Graph',fontdict={'fontname':'Comic Sans MS', 'fontsize':10})
```

```python
# For legend the graph for scale 
plt.plot(x,y,'b*:')
plt.legend(['Single element'])
```

```python
# for saving the graph
plt.plot(x,y,'b*:')
plt.savefig('firstgraph.png',dpi=200)
```

```python
# by using csv file 
mobile = pd.read_csv('mobile.csv')
mobile.head()
```

```python
plt.plot(mobile.year,mobile.india,'b.-',label='India')
plt.plot(mobile.year,mobile.china,'r.-',label='China')
plt.plot(mobile.year,mobile.america,'g.-',label='america')
plt.plot(mobile.year,mobile.japan,'y.-',label='japan')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Mobile Price Year wise',fontdict={'fontname': 'Comic Sans MS','fontsize':20})

plt.legend()
plt.savefig('mobileprice.png')
plt.show()
```

```python
bunch = [0,20,40,60,80,100]
x=[10,20,30,40,50,60]
plt.hist(x,bins=bunch, color='Black')
plt.xlabel('Bunch')
plt.ylabel('Data')
plt.legend(["OK"])
plt.savefig('Histogramgraph')
```

```python
#PieChart
y = np.array([25,35,45,55])
name = ["Ajay","Vijay","Suresh","Sanjay"]
plt.pie(y,labels=name)
```

```python
# Explode
nameexplode = [0,0,0,0.2]
plt.pie(y,labels=name,explode=nameexplode)
```

```python
# Shodow
plt.pie(y,labels=name,explode=nameexplode,shadow=True)
plt.savefig('PieChart.png',dpi=300)
```

```python
choosingcolor = ['b','hotpink','black','m']
plt.pie(y,labels=name,explode=nameexplode,shadow=True, colors= choosingcolor)
plt.legend()
plt.savefig('piechartwithdifcolor.png',dpi=300)
```

```python
# Boxplot
test = np.random.normal(200,50,400)
plt.boxplot(test)
```

```python
# Multiple box plot
test1 = np.random.normal(200,50,400)
test2 = np.random.normal(100,90,300)
test3 = np.random.normal(350,80,200)
test4 = np.random.normal(250,20,400)
data=[test1,test2,test3,test4]
plt.boxplot(data)
```
